```python
import theano
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.mlab as mlab
import theano.tensor as T
x_np=np.linspace(0,20,500) 
y_np=np.sin(x_np)+np.random.normal(0,0.4,500)

# Declare Theano symbolic variables
x_th = theano.shared(x_np,name="x_th")
y_th = theano.shared(y_np,name="y_th")
s_th = T.vector("s_th")
sigma=theano.shared(1.,name="sigma")
x0=T.scalar('x0')
#Theano symbolic expression 
gaussian=T.exp(-(x_th-x0)**2/(2*sigma**2)) #to generate a Gaussian
NW_est=T.sum(y_th*s_th)/T.sum(s_th) #to do NW Kernel estimation

#compile correspnding theano functions
generate_Gaussian=theano.function(inputs=[x0],outputs=gaussian)
nadWatson=theano.function(inputs=[s_th],outputs=NW_est)


t_np=np.zeros(x_np.shape)
s_np=np.zeros(x_np.shape)
idx=0
#loop over each point x, fit a gaussian to all the Y's around that x, and do regression
for xtemp in x_np.tolist():
    s_np=generate_Gaussian(xtemp)
    t_np[idx]=nadWatson(s_np)
    idx=idx+1

    
#generate plots    
plt.scatter(x_np, y_np,color='g')
plt.plot(x_np,np.sin(x_np),color='r')
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.title('Scatter plot of Sample points')
plt.show()


plt.plot(generate_Gaussian(x_np[200]))
plt.ylabel('Gaussian centered at x[200]')
plt.xlabel('Index of x')
plt.title('Gaussian around x[200] ')
plt.show()


plt.scatter(x_np,t_np,color='g')
plt.plot(x_np,np.sin(x_np),color='r')
plt.xlabel('x')
plt.ylabel('t=E[y|x]')
plt.title('Underlying Sine wave estimated by Nadaraya-Watson regression')   
plt.show()
```
