Pringle chip data taken from MatplotLib
---

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

n_angles = 36
n_radii = 8

# An array of radii
# Does not include radius r=0, this is to eliminate duplicate points
radii = np.linspace(0.125, 1.0, n_radii)

# An array of angles
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)

# Repeat all angles for each radius
angles = np.repeat(angles[...,np.newaxis], n_radii, axis=1)

# Convert polar (radii, angles) coords to cartesian (x, y) coords
# (0, 0) is added here. There are no duplicate points in the (x, y) plane
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())

# Pringle surface
z = np.sin(-x*y)

fig = plt.figure()
ax = fig.gca(projection='3d')



plt.show()
```

Visualize pringle chip as a scatter plot with and without noise
--

```python
noise_std=0.1
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(x,y,z,color='r',label="Original Points")
#legend = fig.legend(loc='upper center', shadow=True, fontsize='x-large')
#legend.get_frame().set_facecolor('#00FFCC')
#.legend("Original Points")
#ax.plot_trisurf(x, y, z, cmap=cm.jet, linewidth=0.2)
ax.scatter(x,y,z+np.random.normal(0,noise_std,z.shape),color='g')
plt.title('Scatter Plot of Points with and without Gaussian Noise added')
plt.show()
```

Define Gaussian Function
--

```python
import math
#def genGaussian(mat,mu,sigma):
#    return np.exp(-0.5*np.dot(np.dot((mat-mu),np.linalg.inv(sigma)),(mat-mu).T))
def norm_pdf_multivariate(x, mu, sigma):
  pi=np.pi
  size = len(x)
  #print np.linalg.det(sigma)
  if size == len(mu) and (size, size) == sigma.shape:
    det = np.linalg.det(sigma)
    if det == 0:
        raise NameError("The covariance matrix can't be singular")
    
    norm_const = 1.0/ ( math.pow((2*pi),float(size)/2) * math.pow(det,1.0/2) )
    x_mu = np.matrix(x - mu)
    inv = sigma.I        
    result = math.pow(math.e, -0.5 * (x_mu * inv * x_mu.T))
    return norm_const * result 
  else:
    raise NameError("The dimensions of the input don't match")

def mse(A,B):
    return ((A - B) ** 2).mean()
```

Visualize Z-scaling during one iteration of Nadaraya Watson Regression
--

```python
z_noise=z+np.random.normal(0,noise_std,z.shape)
sigma=np.matrix([[0.1,0.0],[0.0,0.1]])
mu=np.array([x[50],y[5]])
x=x.T
y=y.T
p=np.vstack([x,y])
p=p.T
nIdx=p.shape[0]
for idx2 in range(0,nIdx):
        z_pdf[idx2]=norm_pdf_multivariate(p[idx2,:].T, mu, sigma)

        
fig = plt.figure()
ax = fig.gca(projection='3d')
#ax.scatter(x,y,z_pdf,color='r')
plt.title('Gaussian Centered at a single point evaluated at all given x,y')
ax.scatter(x, y, z_pdf, cmap=cm.jet, linewidth=0.2)
#ax.scatter(x,y,z+np.random.normal(0,noise_std,z.shape),color='g')
plt.show()

fig = plt.figure()
ax = fig.gca(projection='3d')
#ax.scatter(x,y,z_pdf,color='r')
plt.title('Gaussian Scaled by all the noisy points evaluated at all given x,y')
ax.scatter(x, y, z_pdf*z_noise, cmap=cm.jet, linewidth=0.2)
#ax.scatter(x,y,z+np.random.normal(0,noise_std,z.shape),color='g')
plt.show()
```

Do NW Regression with isotropic Gaussian Kernels of Varying widths and observe their impact
----

```python
#given each pair x and y evaluate likelihood
noise_std=0.1

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(x,y,z,color='r',label="Original Points")
plt.title('Pringle Chip Original')
plt.show()        

fig = plt.figure()
ax = fig.gca(projection='3d')
ax.scatter(x,y,z+np.random.normal(0,noise_std,z.shape),color='g')
plt.title('Pringle Chip Original+Noise std:%f'%(noise_std))
plt.show()  

for n in [1.0,0.1,0.01]:
    z_noise=z+np.random.normal(0,noise_std,z.shape)
    x=x.T
    y=y.T
    p=np.vstack([x,y])
    p=p.T
    nIdx=p.shape[0]
    sigma=np.matrix([[n,0.0],[0.0,n]])
    t_est=np.zeros(p.shape[0])

    for idx1 in range(0,nIdx):
        mu=p[idx1,:]
        z_pdf=np.zeros(p.shape[0])
        for idx2 in range(0,nIdx):
            z_pdf[idx2]=norm_pdf_multivariate(p[idx2,:].T, mu, sigma)
            num=np.sum(z_pdf*z_noise)
            den=np.sum(z_pdf)
            t_est[idx1]=num/den

    fig = plt.figure()
    ax = fig.gca(projection='3d')
    ax.scatter(x, y, t_est, cmap=cm.jet, linewidth=0.2)
    plt.title('Pringle Chip re-estimated by Nadaraya Watson Regression \n with an isotropic bivariate Gaussian sig=%f MSE=%f'%(n,mse(z,t_est)))
    plt.show()        
    

```
