Illustration of Nadaraya-Watson Kernel Regression
---

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.mlab as mlab
```

```python
x=np.linspace(0,20,500) 
y=np.sin(x)+np.random.normal(0,0.4,500)
plt.scatter(x, y, color='g')
plt.plot(x,np.sin(x),color='r')
plt.xlabel('x')
plt.ylabel('y=sin(x)')
plt.title('Scatter plot of Sample points')
plt.show()
```

```python
mu=x[200] #take a certain point x and make it the mean of the Gaussian window
sigma = 1 # width of the gaussian
s = mlab.normpdf(x,mu,sigma) # generate a Gaussian envelope for all points centered at x[200]
plt.plot(s)
plt.ylabel('Gaussian centered at x[200]')
plt.xlabel('Index of x')
plt.title('Gaussian around x[200] ')
plt.show()

num = s*y #scale the corresponding y's around x[200]
plt.plot(num)
plt.ylabel('y`s scaled by the Gaussian')
plt.xlabel('Index of x')
plt.title('y`s scaled by the Gaussian around x[200]')
plt.show()
```

```python
#Do the above for every value of x
sigma=1
t=np.zeros(x.shape)
idx=0
for mu in x:
    s = mlab.normpdf(x,mu,sigma)
    num = np.sum(s*y) #Nadaraya-Watson numerator term
    den = np.sum(s) # Nadaraya-Watson denominator term
    t[idx] = num/den # store the estimated values
    idx=idx+1
```

```python
plt.scatter(x, t,color='g')
plt.plot(x,np.sin(x),color='r')
plt.xlabel('x')
plt.ylabel('t=E[y|x]')
plt.title('Underlying Sine wave estimated by Nadaraya-Watson regression')
plt.show()
```

```python
x0=x[200]
gex=np.exp(-(x-x0)**2/(2*sigma**2))
plt.plot(gex)
plt.show()
```
